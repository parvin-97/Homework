### 5.Illustrate relationship between Total Bags and Small Bags using scatter plot.

```python
plt.subplots(figsize =(10,8))
plt.xlabel('Total Bags')
plt.ylabel('Small Bags')
plt.title('Scatter Plot')
plt.scatter("Total Bags", "Small Bags", data =df, color = "g")
```

```python
sns.jointplot(x = "Total Bags", y = "Small Bags", data = df, kind = 'scatter')
```

### 6.Construct AveragePrice distribution.

```python
plt.subplots(figsize =(8,8))
             
sns.countplot(df["AveragePrice"], palette = "rainbow")
```

```python
sns.distplot(df["AveragePrice"], bins = 70)
```

### 7.Which type of Avocados are expensive?(with boxplot)

```python
plt.subplots(figsize =(10,8))
sns.boxplot("type", "AveragePrice", data = df)

#Organic type is more expensive
```

### 8.Construct heatmap correlation matrix.

```python
figuresize = plt.subplots(figsize =(10,8))
sns.heatmap(df.corr(), annot = True, linewidths = 1, fmt = "0.2f")
plt.xticks(rotation = 90)
plt.yticks(rotation = 45)
plt.title("Heatmap")
plt.show()
```

### 9.Find relationship between large and small bags using scatterplot.(use jointplot function)

```python
sns.jointplot("Small Bags", "Large Bags", data = df, kind = "scatter")
```

### 10.Find linear relationship between Total Volume and Bags for each avocado type.(use lmplot function)

```python
sns.lmplot(x ="Total Bags", y = "Total Volume", data = df, col = "type")
```
